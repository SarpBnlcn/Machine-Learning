```python
Q1=veri.sepal_width.quantile(0.25)
Q3=veri.sepal_width.quantile(0.75)
IQR=Q3-Q1

Q1, Q3, IQR
```

```python
Ãœst_SÄ±nÄ±r=Q3+1.5*IQR
Alt_SÄ±nÄ±r=Q1-1.5*IQR

Ãœst_SÄ±nÄ±r,Alt_SÄ±nÄ±r
```

```python
yeni_veri = veri[(veri.sepal_width >= Alt_SÄ±nÄ±r) & (veri.sepal_width <= Ãœst_SÄ±nÄ±r)]
yeni_veri
```

```python
print("Yeni_Veri:", yeni_veri.shape)
```

```python
X = yeni_veri.drop('species', axis=1)
y = yeni_veri['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, random_state=0)
```

```python
model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)
```

```python
y_tahmin = model.predict(X_test)
```

```python
print(y_tahmin)
```

```python
print(y_test)
```

```python
cm=confusion_matrix(y_test, y_tahmin)
print(cm)
```

```python
print("DoÄŸruluk OranÄ±:", accuracy_score(y_test, y_tahmin))
print("SÄ±nÄ±flandÄ±rma Raporu:\n", classification_report(y_test, y_tahmin))
print("KarÄ±ÅŸÄ±klÄ±k Matrisi:\n", confusion_matrix(y_test, y_tahmin))
```
